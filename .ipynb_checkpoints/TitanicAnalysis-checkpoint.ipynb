# Supervised learning 
- ## Classification

***

**Classification** is predicting a label from a discrete dataset. (Blaine et al., 2020). Usually, we see examples of binary classification as these are straightforward, meaning it is or is not. In the "discrete" dataset (Blaine et al., 2020), the class labels are either 0 or 1. For instance, we could apply a detection of whether a given email is spam or not (Blaine et al., 2020) by teaching a model based on the labelled data found. multi-class classification is different as it has "more than two"(Grandini, Enrico Bagli and Giorgio Visani, 2020) class labels. Each number in the classifying label represents a "different class"(Grandini, Enrico Bagli and Giorgio Visani, 2020) or outcome. Different types of classification also exist, like multi-label and imbalanced classification; however, for my problem domain, I think I would need to explore further benefits of binary classification.

***

**Regression** is described as a task of estimating a continuous value (Blaine et al., 2020). It is crucial to note that regressions, on their own, reveal associations exclusively between a "dependent variable and a fixed dataset collection of different variables" (Maulud and Abdulazeez, 2020). There are various regression examples; two important ones are linear and non-linear regression. One assumes a linear relationship between the "independent and dependent variables" (Maulud and Abdulazeez, 2020) by applying a linear equation. The relationship can also be described as a straight line. The other assumes the opposite association between "independent and dependent variables" (Maulud and Abdulazeez, 2020) by applying a non-linear equation accommodating complex and curved patterns in data.

# Unsupervised learning
- ## Clustering

***

**Clustering** is a form of unsupervised learning, and it is the method of finding the natural grouping based on the predefined properties of variables in the dataset(Alok and Bradford, 2019). It uses intuitive sense (Alok and Bradford, 2019) to identify a variable to its group. clustering aims to keep similar data points together whilst keeping dissimilar data points separately. however, the model will not know the difference between the categories as the data is unlabeled compared to classification, where the model predicts outcomes based on labelled data. the "clusters or bunches" (Alok and Bradford, 2019) identified by the model will be determined by "the Euclidian distance" (Alok and Bradford, 2019) between them. Euclidian distance measures the "straight line distance between any two points", described as "the square root of the sum of squares of differences between the coordinates"(Alok and Bradford, 2019). It is a standard metric even though others exist, like the "Manhattan distance"(Alok and Bradford, 2019), which is the distance between two points "measured along a right angle to the axes"(Alok and Bradford, 2019) or the "sum of the absolute value of the differences" between the two points(Alok and Bradford, 2019). The model doesn't know the difference between the variables by name or description, but instead, it categorically identifies clusters based on the distance between two points.

# Reinforcement Learning
- ## Decision making

*** 

**Reinforcement Learning** is described as Agents learning from experience. (Russell and Norvig, 2021) This is compared to supervised learning, which learns passively from input data. The method of reinforcement learning can be described as decision-making. The agent in reinforcement learning is described as having an interaction with the world and "periodically receives rewards"(Russell and Norvig, 2021) that affect the behaviour of the agent to determine "how well it is doing"(Russell and Norvig, 2021). These are very different from the other two types of learning described before; unsupervised and supervised learning are trained on continuous and discrete datasets to predict outcomes. Meanwhile, reinforcement learning is training on the responses it receives from the world and making decisions based on the results. an example of this is if an agent is playing chess, "the reward is 1 for winning, 0 for losing and 1/2 for a draw" (Russell and Norvig, 2021). "Markov decision process can be considered as the basis theory" of reinforcement learning named after "Andrey Markov" (Gao, Chen & Mi 2022) a mathematical framework applied to represent decision-making scenarios with outcomes influenced by a combination of randomness and controllable factors.

## 1. Nave Bayes

Nave Bayes is categorised among the "family of simple probabilistic classifiers" within classification algorithms (Xu, 2016). The algorithm adopts a Nave assumption that "all features are independent of each other," streamlining computations and demonstrating effectiveness in practical applications. However, this assumption's limitation is evident in its oversight of "dependencies among attributes" (Zhang, 2004), where another may influence the truth of one attribute. Despite this limitation, the Nave Bayes approach excels in mitigating the challenges posed by the "curse of dimensionality" (Scikit-learn, 2019), mainly when "dependencies among attributes cancel each other out" (Zhang, 2004).

In examining real-world applications such as spam filtering, it becomes apparent that Nave Bayes performs admirably even with a "small amount of training data to estimate necessary parameters" (Scikit-learn, 2019), reducing the risk of overfitting. Overfitting, a concern in larger, more complex models with numerous parameters, is less likely to occur due to the simplicity and computational efficiency of the Nave Bayes method.

This classification technique has proven successful across diverse real-world scenarios, encompassing "text classification, keyphrase extraction, and medical diagnosis" (Xu, 2016). Its efficacy in handling "dependence distribution" contributes to its widespread adoption in addressing complex problems (Xu, 2016).


Nave Bayes stands out for its swift setup, notably described as "extremely fast compared to sophisticated methods" (Scikit-learn, 2019). This stands in contrast to more intricate algorithms like neural networks, which necessitate a longer setup time and a waiting period for results. In my problem domain, leveraging Nave Bayes involves opting for Gaussian Nave Bayes. This choice is motivated by the dataset's adherence to a normal distribution and "continuous features" (Scikit-learn, 2019). The Gaussian Nave Bayes algorithm excels in handling numerical data, aligning seamlessly with the dataset's characteristics.

## 2. Random Forest


Random Forest has been deemed "extremely successful"  in tackling classification and regression problems, rendering it a versatile and advantageous tool across various domains (Biau and Scornet, 2016). This is solely because of the "combination of its components", including "bagging and the classification and regression trees (CART)-split criterion" (Biau and Scornet, 2016). The algorithm operates by combining "several randomised decision trees" and their collective predictions are obtained through "aggregating the predictions through averaging" (Biau and Scornet, 2016). Noted for its "simplicity and effectiveness", Random Forest employs a "sampling fractions of data" approach, producing an average prediction (Biau and Scornet, 2016). This methodology underscores its significance in feature importance, enhancing accuracy for specific supervised learning problems. Recognised for its accuracy, Random Forest is celebrated for performing well even in "small sample sizes and high-dimensional feature spaces" within machine-learning contexts (Biau and Scornet, 2016). 

However, because of the "combination of different components", we find it hard to interpret how the model works, and a reasonable description of the method can be like a "black box flavour of method" (Biau and Scornet, 2016). Additionally, with the combination of these components, bagging is a "computationally intensive procedure" used to "improve unstable estimates" (Biau and Scornet, 2016). From this, we can infer that the training time for random forests may take longer if the datasets are larger.

## 3. Decision tree

A decision tree is an algorithm which "recursively partitions a dataset of records" until all samples in the dataset belong to a "particular class" or output(Jadhav and Channe, 2016). Techniques used for this process include "Depth-first search, greedy search and breadth-first search" which apply rules to expand nodes in the tree accordingly (Jadhav and Channe, 2016). In a decision tree, every node "denotes a test condition of an attribute", and it is structurally organised to have a "root node, internal and leaf nodes"(Jadhav and Channe, 2016). The root node is the first node in the tree and has no parent node, whereas every other node has at least one parent node, and every node can have as many children nodes as necessary. It follows a "divide and conquer technique" similar to the random forest, and each step the decision tree makes "forms a decision rule"(Jadhav and Channe, 2016). 

Decision trees' advantage is that they are "very simple and fast" for the set-up and result in accurate predictions(Jadhav and Channe, 2016). The structure of a tree is "comprehensible", and it supports "incremental learning"(Jadhav and Channe, 2016). compared to other algorithms, it holds "less memory", and like the naive Bayes algorithm, it can "deal with noisy data"(Jadhav and Channe, 2016). finding the "best split attribute" using a variety of different measures like "entropy, gini index and information gain", which is advantageous (Jadhav and Channe, 2016).

The disadvantage of decision trees is that it can take significant time for the training phase, as described by (Jadhav and Channe, 2016), especially for large datasets due to the process of constructing the tree. Replication issues can lead to decision trees representing some concepts in a "complex" way, affecting the interoperability of the structure(Jadhav and Channe, 2016). Finally, decision trees are susceptible to "overfitting", resulting in reduced generalisation performance on new, unseen data (Jadhav and Channe, 2016).

## 4. K-Nearest Neighbors (KNN)

Out of all the algorithms mentioned, KNN is one of the simplest algorithms to implement and understand. The algorithm is quite self-explanatory as it is based on the similarity between data points, and they are in "close vicinity" (Jadhav and Channe, 2016). KNN is a part of a family of "lazy learners", meaning they are "instance-based classifiers" (Jadhav and Channe, 2016). In the context of "lazy learners", the term "lazy" refers to the fact that these algorithms store all "training samples" and refrain from building a "classifier until a new, unlabeled sample" requires classification (Jadhav and Channe, 2016). This makes using KNN advantageous as there is no training phase, which makes it easy to adapt to new data. KNN takes less "computational time" in the training stage than other "eager-learning algorithms" like decision trees, neural networks and Bayes networks (Jadhav and Channe, 2016). 

The performance of the KNN algorithm is affected by the "value of k", meaning if it is too small, then the algorithm may be "vulnerable to overfitting" due to "noise present in the dataset"(Jadhav and Channe, 2016). On the other hand, if the K value is too big, then the algorithm may "misclassify the test sample" due to similar points coinciding in faraway locations "from its neighbourhood" (Jadhav and Channe, 2016). KNN uses a metric called "Euclidian distance" to find these similarities between the data points and makes the conclusion based on the distance between the values (Jadhav and Channe, 2016).

## 5. Neural Networks (MLP)

One of the most widely used "neural network architectures" is the multi-layer perceptron, which is considered to have "lower complexity and ability to produce satisfactory results" for non-linear relationships (Hemalatha and Rani, 2017). Multilayer Perceptrons utilise "supervised learning" for predictive analysis and accurate results (Hemalatha and Rani, 2017). Multilayer perceptrons show strengths in handling "large quantities of training data" and various types of data, showing their advantageous qualities from their ability to handle complex relationships and suitability for large-scale problems. (Walde et al., 2004). However, this can become computationally expensive for deep architectures with many "neurons, layers and iterations" (Scikit-learn.org, 2023). 

The non-convex nature of the loss function in Multi-layer Perceptrons (MLP) with hidden layers introduces "more than local minimum" (Scikit-learn.org, 2023). Consequently, distinct validation accuracies may arise due to "random weight initialisations" variations. Fine-tuning a multilayer perceptron can make it challenging to obtain the required accuracy. To achieve this one must go through a process of "tuning hyperparameters" such as the number of "neurons, layers and iterations" (Scikit-learn.org, 2023). Overfitting is a problem in neural networks and "other machine learning algorithms." it is the process where a model generalises well with new unseen data, and a multilayer perceptron is "susceptible to overfitting" if the model learns the training data too well. (Lawrence and Giles, 2000)

### 10. Fixing missing values from age

***

**There are several methods for handling missing age data in the Titanic dataset when predicting survivors. Here are some of the most common approaches:**

***10.1. Dropping passengers with missing age data:***

- Removing the whole columns where the age is missing 
- This is the simplest method, but it can lead to a loss of information and a reduction in the dataset size.

***10.2. Imputation with mean or median:***
- This involves replacing the missing age values with the mean or median age of the entire dataset.
- This method is simple, but it can introduce bias if the distribution of missing values differs from that of observed values.

***10.3. Imputation with K-Nearest Neighbors (KNN):***
- This method involves finding the k nearest neighbors (i.e., passengers with similar characteristics) for each passenger with missing age data.
- The missing age value is then replaced with the average or median age of these k nearest neighbors.
- KNN is a more sophisticated method than imputation with mean or median, and it can handle nonlinear relationships between variables.


***10.4. Imputation with multiple imputation:***
- This method involves creating multiple imputed datasets, each with a different set of imputed values for the missing age data.
- The results of multiple imputation are then combined to obtain more robust estimates.


***10.5. Using a machine learning model to predict missing age data:***

- A more advanced approach involves training a machine learning model to predict missing age values based on other variables in the dataset.
- This method can be very accurate, but it requires a careful selection of features and a careful model evaluation.

- The best method for handling missing age data will depend on the specific characteristics of the dataset and the desired outcome.
- In general, it is recommended to use a technique that preserves as much information as possible and does not introduce bias.

- For my method, I decided to make use of machine learning to handle the missing data for my passengers' ages using the selected features;
   - Pclass
   - sex
   - sibsp
   - parch
   - fare
- These features made a valuable addition to predicting the 263 non-null items in my age column. 

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import MinMaxScaler


# Separate passengers with missing age data from those with complete age data
passengers_with_missing_age = data[data['age'].isnull()]
passengers_with_complete_age = data[data['age'].notnull()]

# Select features for predicting missing age data
features = ['pclass', 'sex', 'sibsp', 'parch', 'fare']

# Create training and testing sets for the machine learning model
X_train = passengers_with_complete_age[features]
y_train = passengers_with_complete_age['age']

# Train a Random Forest regressor
regressor = RandomForestRegressor()
regressor.fit(X_train, y_train)

# Predict missing age data using the trained model
X_test = passengers_with_missing_age[features]
predicted_age = regressor.predict(X_test)

# Replace missing age values with predicted age values
data.loc[data['age'].isnull(), 'age'] = predicted_age

# Save the updated dataset
data.to_csv('train_with_predicted_age.csv')

```

### Recall

Recall or 'Sensitivity' is the ratio of instances in my model **correctly predicts positive outcomes** to the total of the **actual positives** in the dataset.

- **Recall = TP/(TP/FN)**

It relates to the model's ability to identify all **True Positives** in the dataset. My model had a score of 71% for my recall, meaning the model was able to capture approximately 71% of the **True Positives** existing in the dataset. Recall is sensitive to **False negatives** indicated in the formula, meaning a higher amount of **False negatives** decreases the amount of Recall for my model. other reasons why my recall may be so low are due to incomplete feature representation in the dataset, model complexity or, again, an imbalance in my class labels.

In this phase, the precision-recall curve is constructed by predicting probabilities using the trained Keras model on the training dataset. The model's output probabilities are flattened, and the precision and recall values are computed across different probability thresholds using the precision_recall_curve function from scikit-learn. The precision-recall curve visually depicts the trade-off between precision and recall at varying classification thresholds. This analysis is valuable for understanding the model's performance characteristics across different levels of classification certainty.

```python
from sklearn.metrics import precision_recall_curve


# Predict probabilities using the trained Keras model
y_scores = simple_model.predict(x_train).ravel()

# Calculate precision and recall
precision, recall, threshold = precision_recall_curve(y_train, y_scores)
```

At this stage, a custom function plot_precision_and_recall is defined to visualise the precision and recall values across different probability thresholds. The function takes three parameters: precision, recall, and threshold. Using the matplotlib library, it creates a line plot with precision values represented in red and recall values in blue. The x-axis corresponds to the probability thresholds, while the y-axis shows the precision and recall values. The function includes labels for clarity, such as "precision" and "recall," and adjusts the plot's appearance by specifying line colours, labels, and font sizes. Finally, a figure of specified dimensions is created using plt.figure(figsize=(14, 7)), and the defined function is called to generate the precision-recall curve. This visualisation aids in understanding how the model's precision and recall trade-off at different probability thresholds.

```python
def plot_precision_and_recall(precision, recall, threshold):
    plt.plot(threshold, precision[:-1], "r-", label="precision", linewidth=5)
    plt.plot(threshold, recall[:-1], "b", label="recall", linewidth=5)
    plt.xlabel("threshold", fontsize=19)
    plt.legend(loc="upper right", fontsize=19)
    plt.ylim([0, 1])

plt.figure(figsize=(14, 7))
plot_precision_and_recall(precision, recall, threshold)
plt.show()
```

The model achieved an average precision of 0.8 and an average recall of 0.7. This suggests that the model can identify many positive examples correctly but also needs some positive examples, meaning there are not enough cases where the features led to the survival of passengers on the Titanic. The precision of the model decreases as the threshold is increased. This means the model is likelier to make false positive predictions as the threshold is raised. The model's performance could be improved by collecting more data, balancing the imbalanced class labels, training the model for longer, or using a different machine learning algorithm, which will be explored later in this project.

## 20.6 Evaluation based on Receiver Operating Characteristic(ROC)

In this stage, the Receiver Operating Characteristic (ROC) curve is generated to illustrate the trade-off between true positive rate (TPR) and false positive rate (FPR) at different classification thresholds. The roc_curve function from scikit-learn is used to compute the TPR, FPR, and corresponding thresholds based on the model's predicted scores on the training dataset. Subsequently, a custom function plot_roc_curve is defined to represent the ROC curve visually. The function creates a line plot where the x-axis corresponds to the FPR, the y-axis to the TPR, and a red diagonal line represents the baseline. The resulting curve provides insights into the model's ability to distinguish between positive and negative instances. This graphical representation is essential for evaluating the model's overall classification performance.

```python
from sklearn.metrics import roc_curve
# compute true positive rate and false positive rate
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_train, y_scores)
```

```python
# plotting them against each other
def plot_roc_curve(false_positive_rate, true_positive_rate, label=None):
    plt.plot(false_positive_rate, true_positive_rate, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'r', linewidth=4)
    plt.axis([0, 1, 0, 1])
    plt.xlabel('False Positive Rate (FPR)', fontsize=16)
    plt.ylabel('True Positive Rate (TPR)', fontsize=16)

plt.figure(figsize=(14, 7))
plot_roc_curve(false_positive_rate, true_positive_rate)
plt.show()
```

The model has an AUC of 0.8, which indicates that it is a good classifier. The ROC curve shows that the model has a high precision for low recall values and a high recall for low precision values. This means that the model is better at identifying true positives when there are fewer false positives, and it is better at identifying all true positives when there are more false positives. The ROC curve shows that the model's performance is sensitive to the threshold. This means adjusting the threshold can improve the model's performance.
